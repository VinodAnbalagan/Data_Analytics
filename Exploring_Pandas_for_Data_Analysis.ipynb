```python
def remove_punc(message):
  # Convert message to string first, handling NaN
  message = str(message)
  Test_punc_removed = [char for char in message if char not in string.punctuation]
  Test_punc_removed_join=''.join(Test_punc_removed)
  return Test_punc_removed_join
```

```python
df['verified_reviews'].apply(remove_punc)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
import gensim
from gensim.utils import simple_preprocess
```

```python
stop_words = stopwords.words('english')
stop_words.extend(['Amazon','amazon','alexa','echo','Alexa','Device','Dot','dot'])
```

```python
gensim.utils.simple_preprocess(df['verified_reviews'][0])
```

```python
def preprocess(text):
  text = str(text)
  result =[]
  for token in gensim.utils.simple_preprocess(text):
    if token not in gensim.parsing.preprocessing.STOPWORDS and len(token)>3 and token not in stop_words:
      result.append(token)
  return result
```

```python
df['verified_reviews_nopunc_nostopwords']=df['verified_reviews'].apply(preprocess)
df
```
